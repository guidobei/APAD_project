Si tratta di dati sui "casi_testati" antecedenti al 18-04-2020: una breve ricerca su internet mi permette di scoprire che i "casi_testati" identificano i "tamponi diagnostici" e la differenza tra "tamponi" e "casi_testati" corrisponde ai tamponi di controllo, effettuati sullo stesso soggetto per confermare la guarigione virologica o per altre necessità di ripetere il test. https://www.sanita24.ilsole24ore.com/art/aziende-e-regioni/2020-05-07/coronavirus-cartabellotta-gimbe-regioni-sotto-soglia-tamponi-ministero-fissi-standard-almeno-250-giorno-100mila-abitanti-100846.php?uuid=ADreTwO

Si procede quindi a riempire i valori mancanti con una stima regionale dei "casi_testati" antecedenti al 18-04-2020. Si prega di prenderla come una stima informale, il cui scopo è soltanto quello di riempire valori NaN.

```python
medie_tamponi = casi_testati_tamponi.groupby('denominazione_regione').mean()
medie_tamponi
```

```python
df_regioni_ind['casi_testati/tamponi'] = np.nan
```

```python
for regione in medie_tamponi.index:
    df_regioni_ind['casi_testati/tamponi'][regione] = medie_tamponi[regione]
```

```python
def presentazione(df):
    for i in range(len(df.columns)):
        df.columns = df.columns.str.replace('_', ' ')
        df.columns = df.columns.str.capitalize()
        
    new_names = []    
    for j in range(len(df.index.names)):
        new_name = df.index.names[j]
        new_name = new_name.replace('_', ' ')
        new_name = new_name.capitalize()
        new_names.append(new_name)
    df.index.names = new_names
    return df
```
